# Data PreProcessing

## Loading the Dataset

```python
import pandas as pd
```

```python
df_churn=pd.read_csv("Churn_Clean.csv")
df=df_churn.copy()
```

The Data is Already Cleaned.
Here we are going to process the data for our Machine Learning Model

## Dropping Irrelevant Columns

```python
df=df.drop("CustomerID",axis=1)
```

“The customerID column is removed because it is a unique identifier and does not provide any predictive information for churn analysis.”

```python
df=df.drop("Count",axis=1)
```

“The Count column is dropped because it contains the same value for all rows and does not provide any useful information for churn analysis.”

```python
df=df.drop("Lat Long",axis=1)
```

```python
df=df.drop("Latitude",axis=1)
```

```python
df=df.drop("Longitude",axis=1)
```

“The Latitude and Longitude columns are dropped as they are unique for each customer and do not provide useful information for predicting churn.”

```python
df=df.drop("Tenure_Group",axis=1)
```

```python
df=df.drop("Monthly_Groups",axis=1)
```

“The Tenure_Group and Monthly_Group columns, created for EDA, are dropped before ML preprocessing as they are not original features.”

```python
df.columns
```

```python
df.head()
```

## Label Encoding

```python
binary_col=[col for col in df.columns if df[col].nunique()==2]
print(binary_col)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for col in binary_col:
    df[col]=le.fit_transform(df[col])
```

```python
df[binary_col].head()
```

All binary categorical columns (columns with only two unique values) were converted into numeric form using Label Encoding.

## One-Hot Encoding

```python
multi_class_col=[col for col in df.columns if df[col].nunique()>2]
print(multi_class_col)
```

```python
three_class=[col for col in df.columns if df[col].nunique()==3]
print(three_class)
```

```python
df['Payment Method'].nunique()
```

```python
multi_col=['Multiple Lines','Internet Service','Online Security','Online Backup','Device Protection','Tech Support','Streaming TV','Streaming Movies',
          'Contract','Payment Method']

for col in multi_col:
    if col not in multi_col:
        print(f"Column not found:{col}")
        
df=pd.get_dummies(df,columns=multi_col,drop_first=True)
```

```python
df.columns
```

```python
df.head()
```

```python
df.dtypes
```

```python
bool_col=[col for col in df.columns if(df[col].dtype==bool)]
print(bool_col)
```

```python
for col in bool_col:
    df[col]=df[col].astype(int)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['Total Charges']=df['Total Charges'].fillna(df['Total Charges'].median())
```

```python
df.isnull().sum()
```

```python
df.to_csv("Churn_Processed.csv",index=False)
```
